# [nyaplot](https://github.com/domitry/nyaplot)凄い！

```ruby
require 'nyaplot'
require 'nyaplot3d'
```

```ruby
x=[];y=[];z=[]
-10.step(10, 0.5) do |i|
  -10.step(10, 0.5) do |j|
    x.push(i)
    y.push(j)
    z.push(Math.sin(Math.sqrt(i*i+j*j))/Math.sqrt(i*i+j*j))
  end
end
z.map!{|val| next (val.nan? ? 0 : val)} #(0,0) will be -inf
```

```ruby
plot = Nyaplot::Plot3D.new
plot.add(:wireframe, x, y, z)
plot.show
```
